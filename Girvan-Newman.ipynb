```python
import networkx as nx
from community import community_louvain
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Loading Graph

```python
G=nx.read_edgelist("karateclub.txt")
```

# Number of Nodes

```python
len(G.nodes)
```

# Number of Edges

```python
len(G.edges())
```

# Girvan-Newman Algorithm

```python
g = G.copy().to_undirected()
modularity = []
community = []
while G.number_of_edges() > 0:
    betweenness = nx.edge_betweenness_centrality(G)
    edge = max(betweenness, key=betweenness.get)
    G.remove_edge(*edge)
    if G.number_of_edges() > 0:
        components = nx.connected_components(G)
        idx = 0
        tmp_partition = {}
        for component in components:
            for inner in list(component):
                tmp_partition.setdefault(inner, idx)
            idx += 1
        community.append(tmp_partition)    
        mod = community_louvain.modularity(tmp_partition,G)
        modularity.append(mod)
```

# Graph of modularity for each iteration

```python
x = np.arange(len(modularity))
y = modularity
plt.xlabel('iteration')
plt.ylabel("modularity")
plt.plot(x,y)
plt.show()
```

# Community Structure for which the graph has highest modularity

```python
part = community[np.argmax(modularity)]
values = [part.get(node) for node in g.nodes()]

nx.draw_spring(g, cmap = plt.get_cmap('jet'), node_color = values, node_size=200, with_labels=False)
```
